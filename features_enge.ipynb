```python
dados = dados[dados['dst_port'].isin(dados['dst_port'].value_counts()[:8].index)]
```

```python
dados.drop(['src_ip', 'dst_ip', 'src_port', 'src_mac', 'dst_mac',
       'protocol', 'timestamp','type'],axis=1,inplace=True)
```

```python
#teste=dados[[ 'dst_port', 'bwd_byts_b_avg', 'bwd_pkts_b_avg',
#       'fwd_blk_rate_avg', 'bwd_blk_rate_avg', 'fwd_seg_size_avg',
#       'bwd_seg_size_avg', 'cwe_flag_count', 'subflow_fwd_pkts',
#       'subflow_bwd_pkts', 'subflow_fwd_byts', 'subflow_bwd_byts', 'hash']]
```

```python
dados.head()
```

```python
agrupado = dados.groupby(["hash","dst_port"]).agg(["min","max","mean"])
```

```python
agrupado.shape
```

```python
agrupado.head()
```

```python
agrupado = agrupado.reset_index()
```

```python
agrupado['dst_port'] = agrupado['dst_port'].astype(str)
```

```python
agrupado.tail()
```

```python
agrupado.columns=agrupado.columns.map('_'.join)
```

```python
pivotado = agrupado.pivot_table(index="hash_",columns='dst_port_')
```

```python
pivotado.columns = pivotado.columns.map('_'.join)
```

```python
pivotado
```

```python
list(agrupado.columns[2:])
```

```python
processada = pd.read_csv("DS_Network_Flow_Final_Protocol_avg_min_max.csv")
processada = processada[['hash','malicious']]
```

```python
final = pivotado.merge(processada,how="left",left_on='hash_',right_on="hash")
```
