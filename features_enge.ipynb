```python
import pandas as pd
```

```python
dados = pd.read_csv("DS_Network_Flow_Final.csv")
```

```python
dados.head()
```

```python
dados = dados[dados['dst_port'].isin(dados['dst_port'].value_counts()[:6].index)]
```

```python
pd.crosstab(dados.protocol,dados.dst_port)
```

```python
dados.shape
```

```python
dados.columns 
```

```python
dados.drop(['src_ip', 'dst_ip', 'src_port','src_mac', 'dst_mac',
       'protocol', 'timestamp','type'],axis=1,inplace=True)
```

```python
dados.head()
```

```python
agrupado = dados.groupby(["hash","dst_port"]).agg(["min","max","sum","mean","std","median","count","var","skew"])
```

```python
agrupado.shape
```

```python
agrupado.head()
```

```python
agrupado = agrupado.reset_index()
```

```python
agrupado['dst_port'] = agrupado['dst_port'].astype(str)
```

```python
agrupado.tail()
```

```python
agrupado.columns=agrupado.columns.map('_'.join)
```

```python
pivotado = agrupado.pivot_table(index="hash_",columns='dst_port_')
```

```python
pivotado.columns = pivotado.columns.map('_'.join)
```

```python
pivotado
```

```python
list(agrupado.columns[2:])
```

```python
processada = pd.read_csv("DS_Network_Flow_Final_Protocol_avg_min_max.csv")
processada = processada[['hash','malicious','repack']]
```

```python
final = pivotado.merge(processada,how="left",left_on='hash_',right_on="hash")
```

```python
final
```

```python
final = final.query("repack == 1.0")
```

```python
final['malicious'].value_counts()
```
